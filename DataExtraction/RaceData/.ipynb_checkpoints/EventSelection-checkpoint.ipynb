```python
base_folder = './Dataset/'
year = '2022'
race = 'zandvoort'
sh_fn = base_folder + year + '/' + race + '/session_history.json'
```

```python
def getLapTimeData(sh):
    lap_times = {}
    for driver in sh:
        # d_i = list(sh.keys()).index(driver)
        # sh_sorted = sorted(sh.items(), key=lambda k: k[d_i][0], reverse=True)
        # print(sh_sorted)
        for lap in sh[driver]:
            lap_i = str(sh[driver].index(lap))
            if int(lap_i) < 2:
                continue
            if lap_i not in lap_times.keys():
                lap_times[lap_i] = {}
            lap_time = lap[1]
            pos = lap[0]
            lap_times[lap_i][driver] = [lap_time, pos]
    for lap in lap_times:
        lt_lap = lap_times[lap]
        lt_sorted = sorted(lt_lap.items(), key=lambda k: k[1])
        lap_times[lap] = dict(lt_sorted)
        for driver in lap_times[lap]:
            lap_time, pos = lap_times[lap][driver]
            lap_times[lap][driver] = lap_time
    return lap_times
```

```python
import json

with open(sh_fn, 'r') as sh_file:
    sh = json.loads(sh_file.read())
    lt = getLapTimeData(sh)
    print(lt)
```

```python
# compare two subsequent drivers and check if their
# speed difference and interval is enough to consider
# it as an 'is approaching' event
def isApproaching(cd_laptime, pd_laptime, interval):
    diff_laptime = cd_laptime - pd_laptime
    if diff_laptime > interval and interval < 4:
        return True
    else:
        return False
```

```python
def madePitStop(laptime, prev_laptime, tyre_info, prev_tyre_info):
    if laptime > (prev_laptime + 15):
        return True
    else:
        return False
```
